```python
import pandas as pd

df_heatwave = pd.read_csv('./data/시도별_폭염일수_평균.csv')
df_precipitation = pd.read_csv('./data/시도별_장마_평균_일강수량.csv')
df_highest_temp = pd.read_csv('./data/시도별_생육기_최고기온.csv')
df_yield_cold = pd.read_csv('./data/복숭아_회귀_데이터셋_1단계.csv')
```

```python
df_heatwave.rename(columns={'year':'연도'}, inplace=True)
df_highest_temp.rename(columns={'year':'연도'}, inplace=True)
df_precipitation.rename(columns={'year':'연도'}, inplace=True)
df_yield_cold.rename(columns={'시도별':'시도명'}, inplace=True)

pd.merge(df_yield_cold, df_heatwave, on=['연도', '시도명']) \
  .merge(df_precipitation, on=['연도', '시도명']) \
  .merge(df_highest_temp, on=['연도', '시도명']) \
  .to_csv('./data/복숭아_회귀_데이터셋_2단계.csv', index=False, encoding='utf-8')
```

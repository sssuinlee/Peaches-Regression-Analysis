```python
import pandas as pd

# 로드
df_daylight_hour = pd.read_csv("월별_기온_강수량_일조시간_15~24.csv")

# 시도명 매핑 적용
df_daylight_hour['시도명'] = df_daylight_hour['지점명'].map(sido_map)

# 날짜 필터링 (생육기) - 3월 15일 ~ 9월 10일로 잡음
df_daylight_hour['일시'] = pd.to_datetime(df_daylight_hour['일시'], format='%Y-%m')
start_date = pd.to_datetime('2000-03-15')
end_date = pd.to_datetime('2000-09-10')
df_daylight_hour = df_daylight_hour[
    df_daylight_hour['일시'].apply( lambda x: start_date.replace(year=x.year) <= x <= end_date.replace(year=x.year) )
]

# 연도, 시도별 일조시간 합계 평균 계산
df_daylight_hour['연도'] = df_daylight_hour['일시'].dt.year
df_daylight_hour_grouped = df_daylight_hour.groupby(['연도', '시도명'])['합계 일조시간(hr)'].sum().reset_index()
df_daylight_hour_grouped['관측소 수'] = df_daylight_hour.groupby(['연도', '시도명'])['지점명'].nunique().values
df_daylight_hour_grouped['평균 일조시간(시간)'] = df_daylight_hour_grouped['합계 일조시간(hr)'] / df_daylight_hour_grouped['관측소 수']
df_daylight_hour_grouped['평균 일조시간(시간)'] = df_daylight_hour_grouped['평균 일조시간(시간)'].round(0).astype(int)
df_daylight_hour_grouped
```

```python
df_daylight_hour_grouped = df_daylight_hour_grouped[df_daylight_hour_grouped['연도'] >= 2015].reset_index(drop=True)
df_daylight_hour_grouped
```

```python
sejong_to_add = []
years_to_substitute = [2015, 2016, 2017, 2018]

# 2015년부터 2018년까지 루프를 돌며 세종 데이터 처리
for year in years_to_substitute:
    # 1. 현재 데이터프레임에 해당 연도의 세종 데이터가 있는지 확인
    is_sejong_exists = '세종특별자치시' in df_daylight_hour_grouped[df_daylight_hour_grouped['연도'] == year]['시도명'].values
    
    # 2. 데이터 대체/추가 조건을 설정합니다.
    #    (2015~2017년은 데이터가 누락되어 있으므로 추가, 2018년은 대전 값으로 통일하기 위해 처리)
    if not is_sejong_exists or year == 2018:
        # 해당 연도의 대전 데이터 찾기
        daejeon_data = df_daylight_hour_grouped[
            (df_daylight_hour_grouped['연도'] == year) & 
            (df_daylight_hour_grouped['시도명'] == '대전광역시')
        ]
        
        if not daejeon_data.empty:
            # 대전 데이터로 세종 데이터 생성
            sejong_row = {
                '연도': year,
                '시도명': '세종특별자치시',
                # 대전의 합계 일조시간(hr) 값 사용
                '합계 일조시간(hr)': daejeon_data['합계 일조시간(hr)'].values[0],
                '관측소 수': 1, # 대체이므로 1로 고정
                # 대전의 평균 일조시간(시간) 값 사용
                '평균 일조시간(시간)': daejeon_data['평균 일조시간(시간)'].values[0]
            }
            sejong_to_add.append(sejong_row)

# 3. 기존 데이터프레임에서 2018년 세종 데이터 제거 (중복 방지 및 대전 값으로 통일)
# 2018년 세종 데이터가 원본에 있다면, 새로 추가할 대전 값으로 대체하기 위해 제거합니다.
df_daylight_hour_grouped = df_daylight_hour_grouped[
    ~((df_daylight_hour_grouped['연도'] == 2018) & (df_daylight_hour_grouped['시도명'] == '세종특별자치시'))
].reset_index(drop=True)

# 4. 새로운 세종 데이터를 기존 데이터에 추가
if sejong_to_add:
    df_daylight_hour_grouped = pd.concat([df_daylight_hour_grouped, pd.DataFrame(sejong_to_add)], ignore_index=True)


# 5. 최종 정렬 및 인덱스 초기화
df_daylight_hour_grouped.sort_values(by=['연도', '시도명'], inplace=True)
df_daylight_hour_grouped.reset_index(drop=True, inplace=True)

# 확인 출력 (선택 사항)
print(df_daylight_hour_grouped[df_daylight_hour_grouped['시도명'] == '세종특별자치시'])
print("\n✅ 2015년~2018년 세종특별자치시 데이터 대체가 완료되었습니다.")
```

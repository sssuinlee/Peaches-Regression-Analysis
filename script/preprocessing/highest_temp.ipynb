```python
import pandas as pd

# 로드
df_temp = pd.read_csv('./data/OBS_ASOS_DD_20251124115512.csv', encoding='cp949')

# 시도명 매핑
df_temp['시도명'] = df_temp['지점명'].map(sido_map)

# 날짜 필터링 (생육기) - 3월 15일 ~ 9월 10일로 잡음
df_temp['일시'] = pd.to_datetime(df_temp['일시'], format='%Y-%m-%d', errors='coerce')
start_date = pd.to_datetime('2000-03-15')
end_date = pd.to_datetime('2000-09-10')
df_temp = df_temp[
    df_temp['일시'].apply( lambda x: start_date.replace(year=x.year) <= x <= end_date.replace(year=x.year) )
]

# 연도, 시도별 최고기온 (생육기 기간 동안)
df_temp['year'] = df_temp['일시'].dt.year
highest_temp = df_temp.groupby(['year', '시도명'])['최고기온(°C)'].max().reset_index(name='생육기_최고기온(°C)')
```

```python
sejong_data = {
    'year': 2018,
    '시도명': '세종특별자치시',
    '생육기_최고기온(°C)': highest_temp[
        (highest_temp['year'] == 2018) & (highest_temp['시도명'] == '대전광역시')
    ]['생육기_최고기온(°C)'].values[0]
}

highest_temp = pd.concat([highest_temp, pd.DataFrame([sejong_data])], ignore_index=True)
highest_temp = highest_temp.sort_values(by=['year', '시도명']).reset_index(drop=True)

highest_temp.to_csv('./data/시도별_생육기_최고기온.csv', index=False, encoding='utf-8')
```
